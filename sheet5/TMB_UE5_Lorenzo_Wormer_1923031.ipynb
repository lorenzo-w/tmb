### Observations:

Distance increases over simulation time, which should be the case in a stretching simulation. Apart from some small local minima, the increase in distance over simulation time is also approximately linear in the simulated range, which means that the behaviour of the molecule with respect to the applied stretching force is somewhat harmonic. Past 40 ps of simulation time the curve flattens a bit, which probably means that beyond the 3.2 nm distance at 40 ps the behaviour can't be approximated as harmonic anymore.

### Discussion of results from umbrella sampling with window width 0.1 nm:

There are no significant gaps for this run, also the overlap between adjacent windows' histograms is somewhat acceptable. Some windows overlap a small amount with their neighbour's neighbour, but as you can see in the run with 0.15 nm, those overlaps can't be reduced without introducing additional gaps, so I chose to accept some amount of this unwanted overlap. 

The generated profile shows that the free energy is increasing with increasing distance, which is to be expected, as the molecule gains potential energy through the stretching. The curve flattens a bit between 2 and 2.5 nm, almost creating a local minimum. This could be due to sterical effects reducing the resistance to pulling in that configuration, or due to a relatively larger volume in phase space for that range of distance values (entropic effect). Beyond 3.2 nm, towards the end of the simulated distance range, you can see a sharp increase in free energy. This might just be a result of poor sampling at the edges, but might also be due to an increase in the molecule's resistance to stretching, which would fit the observation of anharmonic behaviour in the distance plot (results/dist.png), as discussed above.
